## Merging the 2 datasets on the column song_id ##

```python
song_dt=pd.merge(song_dt2,song_dt1,on='song_id',how='left')
```

```python
song_dt.head()
```

### edit the data set columns for easy identification ###

```python
song_dt['song']= song_dt['title']+'-'+song_dt['artist_name']
song_dt.head()
```

## Considering only top 20,000 data samples for quicker processing ##

```python
song_dt=song_dt.head(20000)
```

### Data Calculation ###
#### This calculated data is to b fed to the recommendation systems ####

```python
song_count = song_dt.groupby(['song']).aggregate({'listen_count':'count'}).reset_index()
song_count.head()
```

```python
grouped_sum=song_count['listen_count'].sum()
song_count['percentage']=(song_count['listen_count']/grouped_sum)*100
song_count.sort_values(['listen_count','song'],ascending=[0,1])
```

## Popularity recommendation ##

```python
prm =Recommenders.popularity_recommender_py()
```

```python
prm.create(song_dt,'user_id','song')
```

```python
prm.recommend(song_dt['user_id'][10])
```
